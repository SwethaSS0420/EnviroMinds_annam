## Soil Classification Challenge 1
This notebook summarizes the first Soil Classification project, where the task is to classify soil images into four types: Alluvial, Black, Clay, and Red.

## Project Overview
This project was built to solve a multi-class image classification problem using deep learning. The dataset contains labeled images of four soil types. We used transfer learning with EfficientNet-B3 and ensemble techniques to improve efficiency and accuracy.

## Technologies
Python, PyTorch, EfficientNet-B3, OpenCV (CLAHE), scikit-learn, NumPy, Pandas


## Key Features:
1. CLAHE-based contrast enhancement during preprocessing
2. Augmentation to handle class imbalance (flips, rotation, brightness)
3. Transfer learning with EfficientNet-B3
4. Stratified K-Fold Cross-Validation (5-fold)
5. Ensemble of 5 models for final prediction
6. Metrics: Accuracy, F1 Score

## Why This Approach?
1. Soil images vary greatly in lighting. CLAHE helps enhance texture features.
2. Data augmentation increases diversity and reduces class imbalance issues.
3. EfficientNet-B3 provides a great trade-off between speed and accuracy.
4. Cross-validation and ensemble models improve reliability and generalization.
5. F1 score was prioritized to handle skewed class distribution better than accuracy alone.
