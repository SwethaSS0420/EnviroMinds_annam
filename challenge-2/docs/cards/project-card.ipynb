## Soil Classification Challenge 2


## Project Overview
This project tackles the Soil Classification Challenge on Kaggle, where the task is to classify whether an image contains soil or not. The main challenge is that the training dataset contains only positive examples (soil images), while the test set contains a mix of soil and non-soil images, making this an anomaly detection problem.


## Technologies
Python, PyTorch, EfficientNet-B3, Isolation Forest, One-Class SVM, scikit-learn, NumPy, Pandas


## Key Features:
1. Deep feature extraction with EfficientNet-B3
2. Ensemble of anomaly detection models
3. Synthetic anomaly generation for validation
4. Threshold tuning for best classification accuracy

## Why This Approach?
* Since we only have positive examples during training, supervised learning is not suitable.
* The model learns to recognize soil images and detect anomalies (non-soil).
* Synthetic anomalies are generated to validate and tune the model.
* Visual features are extracted using a pretrained EfficientNet-B3 model.
* An ensemble of Isolation Forest and One-Class SVM models is trained for robust anomaly detection.
